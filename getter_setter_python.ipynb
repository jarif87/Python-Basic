```
class Person:

    def __init__(self,age=0):
        self.age=age
    #setter method
    def set_age(self,age):
        self.age=age
    
    #getter method
    def get_age(self):
        return self.age
```

```
fariha=Person()
```

```
fariha.set_age(20)
```

```
fariha.get_age()
```
